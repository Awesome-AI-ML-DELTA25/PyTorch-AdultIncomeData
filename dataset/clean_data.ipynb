```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('adult.csv')
df
```

```python
percent_missing = (df.isin(['?']).sum(axis=0) / df.apply(len)) * 100
print(percent_missing)

```

```python
df = df[(df['workclass'] != '?') & (df['occupation'] != '?') & (df['native.country'] != '?')]
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df.loc[:, 'income'] = df['income'].apply(lambda x: 1 if x == '>50K' else 0)
df.loc[:, 'sex'] = df['sex'].apply(lambda x: 1 if x == 'Male' else 0)
categorical_columns = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'native.country']
numeric_columns = [i for i in df.columns if i not in categorical_columns + ['income', 'sex']]
```

```python
df
```

```python
df.to_csv('adult_cleaned.csv', index=False)
```
